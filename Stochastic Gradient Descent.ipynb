```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline
```

```python
def f(x, y):
    return 3*x**2 + y**2 + x*y + x + y
```

```python
u = np.linspace(-6,6,100)
x, y = np.meshgrid(u,u)
z = f(x,y)
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
plt.show()
```

```python
fig, ax = plt.subplots()
ct = ax.contour(x, y ,z)
ax.clabel(ct, inline=True, fontsize=10)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x, y, z, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

## Finding the minimum

```python
def sgd(x, y):
    fx = 6*x + y + 1
    fy = 2*y + x + 1
    return fx, fy
```

```python
x = 0.5
y = 0.5
learning_rate = 0.01
i = 1
i_max = 500

while i < i_max:
    gradient_vector = sgd(x, y)
    
    x = x - learning_rate * gradient_vector[0]
    y = y - learning_rate * gradient_vector[1]
    
    i += 1
    print(x, y, i)
```
